```python

import pandas as pd
df = pd.read_excel('/content/Combined_2014_to_2023(1).xlsx')

# calculate and print the number of missing values in each column
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Remove the columns 'tsun' and 'wdir'
df.drop(columns=['tsun', 'wdir','wpgt'], inplace=True)

#check the remaining columns
print(df.columns)
```

```python
# Filling missing 'tavg' values with the average of 'tmin' and 'tmax'
df['tavg'] = df['tavg'].fillna((df['tmin'] + df['tmax']) / 2)

# Check if the operation was successful
print(df['tavg'].isnull().sum())
```

```python
# Remove rows where 'tavg' has missing values
df = df.dropna(subset=['tavg'])

print(df.shape)
```

```python
# Calculate the median values for 'prcp', 'snow', and 'wspd'
median_prcp = df['prcp'].median()
median_snow = df['snow'].median()
median_wspd = df['wspd'].median()
median_pres = df['pres'].median()

# Filling missing values in these columns with the calculated medians
df['prcp'] = df['prcp'].fillna(median_prcp)
df['snow'] = df['snow'].fillna(median_snow)
df['wspd'] = df['wspd'].fillna(median_wspd)
df['pres'] = df['pres'].fillna(median_pres)


print(df[['prcp', 'snow', 'wspd','pres']].isnull().sum())
```

```python
# Saving the DataFrame to an Excel file
df.to_excel('Final.xlsx', index=False)

print("The cleaned dataset has been saved as 'Final.xlsx'.")
```
